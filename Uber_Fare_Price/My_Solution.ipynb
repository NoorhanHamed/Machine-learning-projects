```python
from geopy.distance import great_circle
```

```python
def get_distance(dff):
    try:
        pickup_geo = (dff["pickup_latitude"] , dff["pickup_longitude"]  )
        dropoff_geo = (dff["dropoff_latitude"] , dff["dropoff_longitude"] )
        distance = great_circle(pickup_geo , dropoff_geo).kilometers
        return distance 
    except :
        return np.nan
```

```python
df["distance"] = df.apply(get_distance , axis = 1 )
```

```python
x = df[["Year" , "distance"]]
y = df["fare_amount"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test ,y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=7)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train[x_train.columns] = scaler.fit_transform(x_train[x_train.columns])
x_test[x_test.columns] = scaler.transform(x_test[x_test.columns])
```

```python
from sklearn.linear_model import LinearRegression , Ridge
```

```python
model = LinearRegression()
```

```python
model.fit(x_train , y_train)
```

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print("r2 train is " ,r2_score(y_train , y_train_pred) )
print("r2 test is " ,r2_score(y_test , y_test_pred) )
```

```python
model_ridge = Ridge()
```

```python
model_ridge.fit(x_train , y_train)
```

```python
y_train_pred = model_ridge.predict(x_train)
y_test_pred = model_ridge.predict(x_test)
```

```python
print("r2 train is " ,r2_score(y_train , y_train_pred) )
print("r2 test is " ,r2_score(y_test , y_test_pred) )
```
